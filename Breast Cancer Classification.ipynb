## **Introduction:**

Breast cancer is one of the most prevalent and life-threatening forms of cancer affecting women worldwide. Early detection plays a pivotal role in improving treatment outcomes and increasing the chances of survival. With advancements in medical imaging and machine learning, there has been a growing interest in developing accurate and efficient methods for the early diagnosis of breast cancer.

This project focuses on the analysis and classification of breast cancer tumors using the Breast Cancer Wisconsin (Diagnostic) dataset. The dataset provides a comprehensive set of features extracted from digitized images of fine needle aspirates (FNAs) of breast masses. These features encompass various characteristics of cell nuclei, including size, shape, and texture.

Our primary goal is to build a predictive model that can classify tumors as either malignant or benign based on these features. The model's ability to distinguish between malignant and benign tumors can aid medical professionals in making informed decisions about further diagnostic procedures and treatment strategies.

Throughout this project, we will explore the dataset, preprocess the data, and apply machine learning techniques to train a classification model. We will evaluate the model's performance using relevant metrics and visualize its interpretability. The insights gained from this study not only contribute to the field of medical image analysis but also have the potential to enhance early breast cancer detection, ultimately impacting patient outcomes.

By leveraging machine learning algorithms, we aim to provide a valuable tool for healthcare professionals to assist in the early identification of breast cancer, thereby contributing to the ongoing efforts in the fight against this prevalent and impactful disease.


### Dataset Information:

The dataset consists of 569 instances with 30 numeric features each.
The features include various measurements related to the characteristics of cell nuclei present in breast mass images, such as mean radius, texture, perimeter, area, smoothness, compactness, concavity, concave points, symmetry, and fractal dimension.

### Target Classes:

The target variable is binary, with classes labeled as 'malignant' (0) and 'benign' (1).
Class Distribution: 212 instances of malignant tumors and 357 instances of benign tumors.

### Data Characteristics:

The dataset provides information on the mean, standard error, and worst (or largest) values of the features, resulting in a total of 30 features for each instance.

### Feature Names:

The feature names are provided, and they include descriptors like 'mean radius,' 'mean texture,' 'worst concavity,' etc.

### Missing Values:

The dataset indicates that there are no missing attribute values, which is good for building a reliable model.

Dataset Creator and References:

The dataset was created by Dr. William H. Wolberg, W. Nick Street, and Olvi L. Mangasarian.
References to relevant publications are provided for further understanding.

## Benign:

Benign tumors are non-cancerous growths. They tend to grow slowly and do not invade nearby tissues or spread to other parts of the body. In the context of breast cancer, a benign tumor does not pose a significant threat to health, and its removal may be recommended in some cases for various reasons, such as to relieve symptoms or concerns.

## Malignant:

Malignant tumors, on the other hand, are cancerous growths. They can invade nearby tissues and have the potential to spread (metastasize) to other parts of the body. Malignant breast tumors are the ones that are of primary concern in breast cancer diagnosis and treatment. Early detection of malignant tumors is crucial for better treatment outcomes.

# Separating the features and target 

```python
x =bc.drop(columns = 'label' , axis = 1 )
y = bc['label']
```

```python
print(x)
```

```python
print(y)
```

```python
# Splitting train and test 
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_std = scaler.fit_transform(x_train)

x_test_std = scaler.fit_transform(x_test)
```

```python
print(x_train.shape,x_test.shape)
```

# Building Neural network

```python
import tensorflow as tf 
```

```python
tf.random.set_seed(3)
```

```python
from tensorflow import keras
from keras.layers import Dropout
```

```python
# giving layers of neural network 

model = keras.Sequential([])
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(20, activation= tf.nn.relu))
model.add(tf.keras.layers.Dense(256, activation= tf.nn.relu))
model.add(tf.keras.layers.Dense(256, activation= tf.nn.relu))


model.add(Dropout(0.2))

model.add(tf.keras.layers.Dense(30, activation= tf.nn.softmax)) # output layer
adam = tf.keras.optimizers.Adam(learning_rate = 0.001)
model.compile(optimizer = adam , loss = 'sparse_categorical_crossentropy' , metrics = ['accuracy'])

# adam --> gradient descent
```

```python
model.fit(x_train,y_train,validation_split=0.1,epochs=10)
```

```python
history = model.fit(x_train,y_train,validation_split=0.1,epochs=10)
```

```python
model.fit(x_train_std,y_train,validation_split=0.1,epochs=10)
```

```python
df_exm = pd.DataFrame(model.history.history)
df_exm
```

```python
# visulaization accuracy and loss
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epocs')
plt.legend(['train','val'])
```

```python
plt.figure(figsize =(10,6))
plt.plot(df_exm.loss, marker = '*' ,color = 'red',label = 'Train')
plt.plot(df_exm.val_loss, marker = '.', color = 'green',label = 'Validation')
plt.legend()
plt.grid()
```

```python
plt.figure(figsize =(10,6))
plt.plot(df_exm.accuracy, marker = '*' ,color = 'red',label = 'Train')
plt.plot(df_exm.val_accuracy, marker = '.', color = 'green',label = 'Validation')
plt.legend()
plt.grid()
```

```python
pred_test = model.predict(x_test_std)
```

```python
pred_final =  pred_test.argmax(axis =1)
```

```python
from sklearn.metrics import confusion_matrix
tab = confusion_matrix(y_test , pred_final)
tab
```
